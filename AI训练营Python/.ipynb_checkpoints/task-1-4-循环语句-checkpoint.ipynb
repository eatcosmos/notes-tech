## 1.while 循环
循环和条件语句正好相互补充，  
循环不间断，条件有间断，  
循环有次数，条件仅一次，  
循环是时间，条件是交互，  
可以写入str、list或任何序列，长度非零则视为真值，执行循环体

````python
count = 0 # 有次数所以用循环
while count < 3:
    temp = input("猜猜大叔想的是哪个数字？")
    guess = int(temp)
    if guess > 8:
        print("大了，大了，你真笨！")
    else:
        if guess == 8:
            print("你太了解大叔的心思了！")
            print("哼，猜对了也没有奖励！")
            count = 3 # 提前结束下一次循环
        else:
            print("小了，小了")
    count = count + 1
print("Game Over!")
````

````python
# 可以写入str、list或任何序列，长度非零则视为真值，执行循环体；否则视为假值，不执行循环体。
string = 'abcd' # 结尾不包含\0，注意和c/cpp区别
print(len(string))
while string:
    print(string)
    string = string[1:]
    print(string)
    print(len(string))
````

## while-else 循环
等同于 while-if-else
```python
while 布尔表达式:
    代码块
else:
    代码块
```

````python
count = 0
while count < 5:
    print("%d is less than 5" % count)
    count = count + 1
else:
    print("%d is not less than 5" % count)
````

````python
count = 0
while count < 5:
    print("%d is less than 5" % count)
    count = 6 # 适用于非常规结束
    break
else:
    print("%d is not less than 5" % count)
````

## for 循环
通用的序列迭代器，序列包括有序序列或者可迭代对象

````python
for i in 'ILoveSGO':
    print(i, end=' ') 
````

````python
member = ['张三', '李四', '刘德华', '刘六']
for each in member:
    print(each)
    
````

````python
for i in range(len(member)): # range索引化整型数
    print(i)
    print(member[i])
````

````python
# 遍历任何可迭代对象
set1 = {1,2,3}
for v in set1.items():
    print(v)
````

````python
# 遍历任何可迭代对象.items()
dic = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
for key, value in dic.items():
    print(key, value, sep=':', end=' ')
    print(key, value, end=' ') # 默认sep=' '
````

````python
# 遍历任何可迭代对象.items()
dic = {'a': 1, 'b': 2, 'c':3, 'd': 4}
for key in dic.keys():
    print(key, end=' ')
````

## 4.for - else 循环
```python
for 迭代变量 in 可迭代对象:
    代码块
else:
    代码块
```

````python
print(10 / 3)
print(10 // 4) # 向下取整
print(10 // 4.3) # 向下取整，同时增加小数点
````

range(索引开始, 如果索引是0这个位置正好是长度)  
range([start,] stop[, step=1])

````python
for i in range(2, 5):
    print(i)
````

````python
for i in range(0, 5):
    print(i)
````

````python
for i in range(5):
    print(i)
````

````python
for num in range(10, 20):
    for i in range(2, num):
        if num % i == 0: # 确定余数
            j = num / i  # 没有余数说明可以整除
            print('%d 等于 %d * %d' % (num, i, j))
            break
    else:
        print(num, '是一个质数')
````

## enumerate() 函数 可以理解为某种有序序列
enumerate(sequence, [start=0])

- sequence: 一个序列、迭代器或其他支持迭代对象
- start: 下标起始位置
- 返回枚举对象

````python
seasons = ['Spring', 'Summer', 'Fall', 'Winter']
print(enumerate(seasons))
````

````python
seasons = ['Spring', 'Summer', 'Fall', 'Winter']
lst = list(enumerate(seasons))
print(lst)
lst = list(enumerate(seasons, start=1))
print(lst)
````

for i, a in enumerate(A):  
    do something with a  
可以理解为特定有序序列的泛化函数，但是多了一个索引key  

````python
languages = ['Python', 'R', 'Matlab', 'C++']
for language in languages:
    print('I love', language)
print('Done!')

for i, language in enumerate(language, 2):
    print(i, 'I love', language)
print('Done!')

for i, language in languages:
    print(i, 'I love', language)
print('Done!')
````

## break 语句
特别适用于无限循环的满足条件退出  
哦，适合循环的满足条件退出  

````python
import random
secret = random.randint(1, 10) # [1, 10]之间的随机数

while True:
    temp = input("猜一猜小姐姐想的是哪个数字？")
    guess = int(temp)
    if guess > secret:
        print("大了, 大了")
    else:
        if guess == secret:
            print("你太了解小姐姐的心思了！")
            break
        else:
            print("小了， 小了！")
print("结束")
````

## continue 语句

````python
for i in range(10): # 0~9
    if i % 2 != 0: # 求余数
#         print(i)
        continue
    i += 2 # 最终输出范围中的偶数
    print(i)
````

## pass 语句
pass 语句的意思是“不做任何事情”  
`pass`是空语句，不做任何操作，只起到占位的作用，其作用是为了保持程序结构的完整性。尽管`pass`语句不做任何操作，但如果暂时不确定要在一个位置放上什么样的代码，可以先放置一个`pass`语句，让代码可以正常运行。

## 推导式
**列表推导式**
就是自动根据符合条件元素公式生成列表
```python
[ expr for value in collection [if condition]]
```

````python
x = [2, 3, 0, 3, 3]
y = [a * 2 for a in x]
print(y)
````

````python
x = [(i, i ** 2) for i in range(6)]
print(x)
````

````python
x = [i for i in range(100) if (i % 2) != 0 and (i % 3) == 0]
print(x)
````

````python
# 并列
a = [(i, j) for i in range(0, 3) for j in range(0, 3)]
print(a)
````

````python
x = [[i, j] for i in range(0, 3) for j in range(0, 3)]
print(x)

x[0][0] = 10
print(x)

x[2][0] = 20
print(x)

x[0][2] = 10
print(x)
````

## 元组推导式
(expr for value in collection [if condition])

````python
a = (x for x in range(10))
print(a)
print(tuple(a))
````

## 字典推导式
{key_expr: value_expr for value in collection [if condition]}

````python
b = {i: i % 2 == 0 for i in range(10) if i % 3 == 0}
print(b)
# {0: True, 3: False, 6: True, 9: False}
````
