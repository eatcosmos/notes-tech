## 4. try - except - finally 
```python
try:  
      
except Exception[as reason]:  
        
finally:    
       
```

`try``finally`  

`try``except``finally`  


````python
def divide(x, y):
    try:
        result = x / y # 
        print('result is', result)
    except ZeroDivisionError:
        print('division by zero!')
    finally: # else
        print("executing finally clause")

divide(2, 1)
divide(2, 0)
divide('2', '1')
````

````python
def divide(x, y):
    try:
        result = x / y # 
        print('result is', result)
    except ZeroDivisionError:
        print('division by zero!')
    print("executing finally clause") # 

divide(2, 1)
divide(2, 0)
divide('2', '1')
````
