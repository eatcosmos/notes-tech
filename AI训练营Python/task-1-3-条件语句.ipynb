## 3.if - elif -else
if - if 用于第一个if之后不需要执行条件操作

```python
# 照抄
temp = input('请输入成绩：') # 条件语句在互动输入中特别有用
source = int(temp)
if 100 >= source >= 90: # 注意，比较大小两边都可以的
    print('A')
elif 90 > source >=80:
    print('B')
elif 80 > source >=60:
    print('C')
elif 60 > source >= 0:
    print('D')
else:
    print('输入错误！')
```
