3.字符串的常用内置方法
- capitalize()将字符串的第一个字符->大写  
【例子】

```python
str2 = 'xiaoxie'
print(str2.capitalize())
```

```python
# lower() upper() swapcase()
str2 = "DAXIExiaoxie"
print(str2.lower())
print(str2.upper())
print(str2.swapcase())
```

```python
str2 = 'DAXIExiaoxie'
print(str2.count('xi')) # count(str, beg=0, end=len(string))
```

- `endswith(suffix, beg=0, end=len(string))` 检查字符串是否以指定子字符串 `suffix` 结束，如果是，返回 True，否则返回 False。如果 `beg` 和 `end` 指定值，则在指定范围内检查。
- `startswith(substr, beg=0,end=len(string))` 检查字符串是否以指定子字符串 `substr` 开头，如果是，返回 True，否则返回 False。如果 `beg` 和 `end` 指定值，则在指定范围内检查。

【例子】

```python
str2 = "DAXIExiaoxie"
print(str2.endswith('ie'))
print(str2.endswith('xi'))
print(str2.startswith('Da'))
```

- `find(str, beg=0, end=len(string))` 检测 `str` 是否包含在字符串中，如果指定范围 `beg` 和 `end`，则检查是否包含在指定范围内，如果包含，返回开始的索引值，否则返回 -1。
find 作用和 in有点像，似乎是in的在字符串中的扩展
- `rfind(str, beg=0,end=len(string))` 类似于 `find()` 函数，不过是从右边开始查找。

【例子】

```python
str2 = "DAXIExiaoxie"
print(str2.find('xi'))
print(str2.find('ix'))
print(str2.rfind('xi'))
```

- `isnumeric()` 如果字符串中只包含数字字符，则返回 True，否则返回 False。

【例子】
其实所谓方法就是我们日常业务任务的总结呀

```python
str3 = '12345'
print(str3.isnumeric()) # True，就是各种内置方法，当然了你也可以自己添加新的方法啦
str3 += 'a'
print(str3.isnumeric()) # False
```

其实一切都是为了处理协议，只是这里的协议是零散的  
- `ljust(width[, fillchar])`返回一个原字符串左对齐，并使用`fillchar`（默认空格）填充至长度`width`的新字符串。
- `rjust(width[, fillchar])`返回一个原字符串右对齐，并使用`fillchar`（默认空格）填充至长度`width`的新字符串。

【例子】

```python
str4 = '1101'
print(str4.ljust(8, '0'))
print(str4.rjust(8, '0'))
```

- `lstrip([chars])` 截掉字符串左边的空格或指定字符。
- `rstrip([chars])` 删除字符串末尾的空格或指定字符。
- `strip([chars])` 在字符串上执行`lstrip()`和`rstrip()`。

【例子】

```python
str5 = '   I Love LsgoGroup  '
print(str5)
print(str5.lstrip()) 
print(str5.lstrip().strip('I'))
print(str5.rstrip())
print(str5.strip())
print(str5.strip().strip('p'))
print(str5.strip().strip('u'))
```

- `partition(sub)` 找到第一个子字符串sub，把字符串分为一个三元组`(pre_sub, sub, fol_sub)`，如果字符串中不包含sub则返回`('原字符串','','')`。
- `rpartition(sub)`类似于`partition()`方法，不过是从右边开始查找。

【例子】

```python
str5 = '  I Love LsgoGroup '
print(str5.strip().partition('o')) # .有点类似管道的意味
print(str5.strip().partition('m')) # 现实世界是复杂多变得
print(str5.strip().rpartition('o')) # 返回的是一个元组
```

- `replace(old, new [, max])` 把 将字符串中的`old`替换成`new`，如果`max`指定，则替换不超过`max`次。

【例子】

```python
str5 = ' I Love LsgoGroup  '
print(str5.strip().replace('I', 'We')) # 貌似strip()编程标配了
```

- `split(str="", num)` 不带参数默认是以空格为分隔符切片字符串，如果`num`参数有设置，则仅分隔`num`个子字符串，返回切片后的子字符串拼接的列表。

【例子】

```python
# 这个很强大，不用每次都到网上搜了哈，总之就是处理数据是够了哈
str5 = '  I Love LsgoGroup  '
print(str5.strip().split()) # 默认以空格为分割符切片字符串
print(str5.strip().split('o'))
```

【例子】

```python
u = 'www.baidu.com.cn' # 这个在爬虫里用的比较多吧
print(u.strip())

print(u.split('.')) # 分隔所有的
print(u.split('.', 0)) # 不分隔
print(u.split('.', 1)) # 分隔1次
print(u.split('.', 2)) # 分割2次

print(u.split('.', 2)[1]) # 

u1, u2, u3 = u.split('.', 2) # 有点类似解包了
print(u1)
print(u2)
print(u3)
```

## 4.字符串格式化
- format格式化函数，可以理解为一种翻译吧

```python
str8 = '{0} Love {1}'.format('I', 'Lsgogroup') # 位置参数
print(str8)

str8 = '{a} Love {b}'.format(a='I', b='Lsgogroup') # 关键字参数
print(str8) # I Love Lsgogroup

str8 = '{0} Love {b}'.format('I', b='Lsgogroup') # 混合使用的时候位置参数需要在关键字参数之前
print(str8)

str8 = '{0:.2f}{1}'.format(27.658, 'GB') # 带精度属性，保留小数点后两位
print(str8)
```

前面是字符串位置格式化，不能转换格式  

- Python 字符串格式化符号


 符   号 | 描述
:---:|:---
%c | 格式化字符及其ASCII码
%s | 格式化字符串，用str()方法处理对象
%r | 格式化字符串，用rper()方法处理对象，能够重现它所代表的对象
%d | 格式化整数
%o | 格式化无符号八进制数
%x | 格式化无符号十六进制数
%X | 格式化无符号十六进制数（大写）
%f | 格式化浮点数字，可指定小数点后的精度
%e | 用科学计数法格式化浮点数
%E | 作用同%e，用科学计数法格式化浮点数
%g | 根据值的大小决定使用%f或%e
%G | 作用同%g，根据值的大小决定使用%f或%E


【例子】

```python
str = '%c' % 97 
print(str)
# print({0:%c}.format(97)) # SyntaxError: invalid syntax
```

```python
print('%c' % 97) # a
print('%c %c %c' % (4, 5, 6))
print('%d + %d = %d' % (4, 5, 9)) # 格式化整数
print("我叫 %s 今年 %d 岁!" % ('小明', 10))  # 我叫 小明 今年 10 岁!
print('我叫 %s 今年 %d 岁!' % ('小明', 10)) # 注意是英文的逗号
print('%o' % 10) # 先转换为8进制，再以字符串形式打印之
print('%x' % 10) # 十六进制
print('%X' % 10) # 大写
print('%f' % 27.434)
print('%e' % 43.323) # 科学计数法
print('%E' % 32.32324) 
text = 'I am %d years old.' % 22
print('I said: %s.' % text)
print('I said: %r.' % text)

import datetime
d = datetime.date.today()
print('%r' % d)
```

- 格式化操作符辅助指令

符号 | 功能
:---:|:---
`m.n` | m 是显示的最小总宽度,n 是小数点后的位数（如果可用的话）
`-` | 用作左对齐
`+` | 在正数前面显示加号( + )
`#` | 在八进制数前面显示零('0')，在十六进制前面显示'0x'或者'0X'(取决于用的是'x'还是'X')
`0` | 显示的数字前面填充'0'而不是默认的空格

【例子】

```python
print('%5.1f' % 23.221) # 最小总宽度
print('%6.2f' % 23.221) # 小数点是不会截去的
print('%10d' % 10) # 用空格填充
print('%-10d' % 10) # 右边填充空格
print('%+d' % 10) # +10
print('%#o' % 10) # 0o12
print('%010d' % 5) # 0000000005
```
