## 直接赋值：就是创建对象的引用（别名）
id(变量) 表示变量所指向对象的内存地址

```python
tuple1 = (1, 2, 3, 4, 5, 6, 7 ,8)
tuple2 = tuple1[:] # 对于元组，tuple[:] 这种方式其实就是直接赋值
print(id(tuple1) == id(tuple2)) # 父对象内存地址相等
print(id(tuple1) == id(tuple2)) # 子内存地址相等

tuple3 = tuple1 # 直接赋值
print(id(tuple1) == id(tuple3)) # 父对象内存地址相等
```

```python
# 元组的浅拷贝无法用tuple[:]形式，如何实现？
tuple1 = (1, 2, 3, 4, 5, 6, 7 ,8) # 里面的子对象类型是<class 'int'>
print(tuple1[0], type(tuple1[0]))
tuple2 = tuple() # 创建之后()不能再改了
print(tuple1, type(tuple1))

# tuple2.extend(tuple1) # AttributeError: 'tuple' object has no attribute 'extend'
# print(tuple2, type(tuple2))

tuple2 = (tuple1[0], tuple1[1], tuple1[2], tuple1[3], tuple1[4], tuple1[5], tuple1[6], tuple1[7])
print(id(tuple1) == id(tuple2)) # 相当于拷贝了父对象
print(id(tuple1[0]) == id(tuple2[0])) # 子对象相同

tuple2 = (t for t in tuple1) # 元组生成式的效果就是浅拷贝，和上面把tuple1[index]用于初始化tuple2效果一样
tuple2 = tuple(tuple2)
print(id(tuple1) == id(tuple2))
print(id(tuple1[0]) == id(tuple2[0])) 
```

## 浅拷贝：拷贝父对象，不会拷贝对象的内部的子对象
import copy # 引入copy模块

```python
list1 = [1, 2, 3, 4, 5, 6, 7 ,8]
list2 = list1[:] # 对于列表list[:] 这种方式是浅拷贝：
print(id(list1) == id(list2)) # 父对象的内存地址不相等
print(id(list1[0]) == id(list2[0])) # 内部子对象的内存地址相等
print(id(list1[0]) == id(list2[0])) # 内部子对象的内存地址相等
```

```python
# 元组的全切片是引用
tuple1_slice1 = (1,2)
tuple1_slice2 = (3,4)
tuple1_slice3 = (5,6)
tuple1 = tuple1_slice1 + tuple1_slice2 + tuple1_slice3
print(id(tuple1[2]) == id(tuple1_slice2[0]))
print(id(tuple1[2:4]) == id(tuple1_slice2)) # False 部分切片是浅拷贝

print(id(tuple1[0:]) == id(tuple1)) # True
print(id(tuple1[:]) == id(tuple1)) # True 全切片是引用，地址不变，指向相同的内存位置
```

```python
# 列表的全切片是浅拷贝
list1 = [1,2,3,4,5,6]
print(id(list1[:]) == id(list1)) # False 和元组区别
```

```python
tuple1 = (1,2,3,)
print(id(tuple1) == id(tuple1[0:])) # True 引用

list1 = [1,2,3]
print(id(list1) == id(list1[0:])) # False 浅拷贝
```

## 深拷贝copy.deepcopy()： 引入copy 模块的 deepcopy 方法，完全拷贝了父对象及子对象

深度拷贝需要引入 copy 模块才行

```python
import copy # import进copy模块
tuple1.copy() # AttributeError: 'tuple' object has no attribute 'copy'
# 错误会导致程序中断，所以加入异常捕获是必要的
tuple2 = copy.deepcopy() 
print(tuple2)
```


- 创建元组可以用小括号 ()，也可以什么都不用，为了可读性，建议还是用 ()。
- 元组中只包含一个元素时，需要在元素后面添加逗号，否则括号会被当作运算符使用。

【例子】


```python
x = (1)
print(type(x))
x = 1, 2
print(type(x))
x = []
print(type(x))
x = ()
print(type(x)) # <class 'tuple'>
x = (1, )
print(type(x))
```


## 2.更新和删除一个元组
【例子】


```python
week = ('Monday', 'Tuesday', 'Thurday', 'Friday')
week1 = week[:2] + ('Wednesday',) + week[2:] # 都是引用的里面的对象进行组合成新对象的
print(week1)
print(id(week[0]) == id(week1[0]))

week = ('Monday', 'Tuesday', 'Thurday', 'Friday')
# week = week[:2] + ('Wednesday') + week[2:] # TypeError: can only concatenate tuple (not "str") to tuple
print(week)
```


元组有不可更改（immutable）的性质，因此不能直接给元组的元素赋值，但是只要元组中的元素可更改 (mutable)，那么我们可以直接更改其元素，注意这跟赋值其元素不同。


```python
t1 = (1, 2, 3, [4, 5, 6])
print(t1)
print(id(t1[3][0]))

t1[3][0] = 9 # 这里是更改的t1元素t1[3]里面的子对象
print(t1)
print(id(t1[3][0])) # 所谓的改就是地址不一样了
```

## 3.元组相关的操作符
- 等号操作符 `==`
- 连接操作符 `+`
- 重复操作符 `*`
- 成员关系操作符 `in`、`not in`

「等号 ==」，只有成员、成员位置都相同时才返回True。

元组拼接有两种方式，用「加号 +」和「乘号 *」，前者首尾拼接，后者复制拼接。

【例子】


```python
t1 = (123, 456)
t2 = (456, 123)
t3 = (123, 456)

print(t1 == t2)
print(t1 == t3)
print(id(t1) == id(t3))

t4 = t1 + t2 # 连接
print(t4)

t5 = t3 * 3 # 重复操作符
print(t5)

t3 *= 3
print(t3)

print(123 in t3)
print(456 not in t3)
```

## 4.元组内置方法
元组大小和内容都不可更改，因此只有查询这一类方法
【例子】


```python
t = (1, 10.31, 'python')
print(t.count('python')) # 统计次数
print(t.index(10.31)) # 查找主键值
```


## 5.解压元组
【例子】解压/解包为一维数组，有几个元素左边扩招定义几个变量


```python
t = (1, 10.31, 'python')
# (a, b) = t # ValueError: too many values to unpack (expected 2)
# print(a, b)
(a, b, c) = t
print(a, b, c)
```


【例子】解压二维数组，按照元组里面的元组结构来定义变量


```python
t = (1, 10.31, ('OK', 'python'))
(a, b, c) = t
print(a, b, c)
```


【例子】如果你只想要元组其中几个元素，用通配符*，英文叫wildcard，在计算机语言中代表0个或者多个元素，存放在列表中


t = 1,2,3,4,5
a,b,c,d,*rest,e = t
print(a,b,c,d,e)
print(rest, type(rest))
